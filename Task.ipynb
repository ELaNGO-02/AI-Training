```python
import pandas as pd
import matplotlib.pyplot as plt

#task
# Load dataset and view first 5 rows.​

# Find total sales across all regions.​

# Identify product with highest profit.​

# Create a bar chart of Sales by Region.​

# Check if dataset has missing values




# Load dataset and view first 5 rows.

salesData = pd.read_csv('Sales.csv')
# print(salesData)
salesData.head()
```

```python
salesData.describe()
```

```python
salesData
```

```python
# Find total sales across all regions.
totalSales = salesData['Sales'].sum()
print(totalSales)
```

```python
# Identify product with highest profit.​
highestProfitIndex = salesData['Profit'].idxmax()
highestProfitProduct = salesData.loc[highestProfitIndex, 'Product']
print(highestProfitProduct)
```

```python
# Create a bar chart of Sales by Region.
salesData.groupby('Region')['Sales'].mean().plot(kind = 'bar')
# plt.ylabel("Sales")
plt.title("Sales By Region")
plt.show()
```
